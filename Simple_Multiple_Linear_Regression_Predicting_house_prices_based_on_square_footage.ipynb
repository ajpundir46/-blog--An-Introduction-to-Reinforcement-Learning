```python

import numpy as np
import zipfile

with zipfile.ZipFile('boston-housing.zip', 'r') as zip_ref:
    zip_ref.extractall('boston-housing')

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the data
train_data = pd.read_csv('/content/boston-housing/train.csv')
test_data = pd.read_csv('/content/boston-housing/test.csv')


test_data.head()
column_names = train_data.columns.tolist()
print(column_names)
# Assuming the columns related to square footage and house prices are named appropriately
X_train = train_data[['zn']]
y_train = train_data['medv']

X_test = test_data[['zn']]
y_test_data = pd.read_csv('/content/boston-housing/submission_example.csv')
y_test = y_test_data['medv']

# Initialize the model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
rmse = np.sqrt(mse)

print("Root Mean Squared Error:", rmse)
```

```python
import matplotlib.pyplot as plt

# Plot the actual vs predicted prices
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', label='Actual vs Predicted')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=3, color='red')  # Plot a diagonal line
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Prices (Linear Regression)')
plt.legend()
plt.grid(True)
plt.show()

```
